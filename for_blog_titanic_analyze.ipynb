## LabelEncoder

```python

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(data['Sex'])
data['Sex'] = le.transform(data['Sex'])

data.head(3)
```

## One-Hot-Encoding

```python
from sklearn.preprocessing import OneHotEncoder

p = data['Sex'].values.reshape(-1, 1)
onehot = OneHotEncoder()
onehot.fit(p)
p = onehot.transform(p)
p = p.todense()

data['female'] = p[:,0]
data['male'] = p[:, 1]

data.head(3)
```
